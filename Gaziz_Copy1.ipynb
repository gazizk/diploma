```python
not_digits_unique_values: tp.Any = set()

for col in df.columns:
    for value in set(df[col]):
        if isinstance(value, str) and not value.isnumeric():
            not_digits_unique_values.add(value)

not_digits_unique_values
```

```python
empty_scoring: tp.List[str] = ["н/я", "нет оценки"]
```

```python
def preprocessing(*, df: pd.DataFrame) -> pd.DataFrame:
    data = df.copy()
    consist_only_nans: tp.List[str] = []

    for col in data.columns:
        if data[col].isna().sum() == data.shape[0]:
            consist_only_nans.append(col)
        
        data[col] = data[col].apply(lambda x: str(x).replace("*", ""))
        data[col] = data[col].apply(lambda x: 0 if x in empty_scoring else x)

    data.drop(consist_only_nans, axis=1, inplace=True)
    data = data.astype(float)
    
    return data


data = preprocessing(df=df)
data
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error


def plot_fold_stats(mean_test_score: tp.List[float], 
                    std_test_score: tp.List[float],
                    title: str = "RSME"
                    ) -> None:
    scores = np.array([(mean, std) for mean, std in zip(mean_test_score, std_test_score)])

    plt.errorbar(
        x=range(len(scores)),
        y=scores[:, 0],
        yerr=scores[:, 1],
        fmt='o',
    )

    plt.ylabel(title)
    plt.show()


def train(*,
          train_columns: tp.List[str], 
          df: pd.DataFrame, 
          target_col: str,
          parameters: tp.Dict[str, tp.Any],
          model = RandomForestRegressor,
          ) -> None:
    X = df.loc[:, train_columns]
    y = df.loc[:, target_col]

    # Drop student who don't take target course
    mask = ~y.isna()
    X = X[mask].copy().reset_index(drop = True)
    y = y[mask]

    # Replace Nan with mean of column
    for col in X.columns:
        X[col].fillna(X[col].median(), inplace=True)
    assert y.isna().sum() == 0, "Nan in targets"
    assert X.isna().sum().sum() == 0, "Nan in X"
    
    # Corvert to numpy
    X, y = X.values, y.values

    # Train and test datasets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.1, random_state=42)

    # Baseline. Predicting with constant value
    # If loss is RMSE -> use mean
    # If loss is MAPE -> use median

    mean_target = np.median(y_train)  # y_train.mean()
    preds = [mean_target] * len(y_test)

    print(f"Baseline: MAE = {mean_absolute_error(y_test, preds)}")

    # Grid Search
    if DEBUG:
        for key in parameters.keys():
            parameters[key] = parameters[key][:2]
    
    estimator = model()
    regressor = GridSearchCV(
        estimator=estimator, 
        param_grid=parameters, 
        n_jobs=-1,
        cv=5,
        scoring="neg_mean_absolute_error"
    )

    regressor.fit(X_train, y_train)


    # Fold statistics
    mean_test_score = regressor.cv_results_["mean_test_score"]
    std_test_score = regressor.cv_results_["std_test_score"]
    plot_fold_stats(-mean_test_score, std_test_score, "MAE")

    # Train on best models
    best_params = regressor.best_params_
    print(f"best params searched by Grid: \n {best_params}")
    
    # Retraining on best params and on whole train dataset
    regressor = model(**best_params)
    regressor.fit(X_train, y_train)

    # Test
    preds = regressor.predict(X_test)
    #print(mean_absolute_error(y_test, preds))
    importance = regressor.feature_importances_
    #print(importance)
    return [mean_absolute_error(y_test, preds), importance]
```
